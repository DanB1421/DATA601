<b>Question 1:</b> [<span style="color: red;">10 points</span>]<br>
Following is the data on how many candies they sold and of what type<br>

| Candy type | Amount sold (in millions) |
| --- | --- |
| Wonka Bar | 6.1 |
| Everlasting Gobstoppers |0.9 |
| Fizzy Lifting Drinks    | 1.0 |
| Hair Toffee             | 0.8 |
| Wonka Gum               | 1.2 |

You need to prepare a pie chart with the lowest performing candy separated (use explode with 0.1 for that slice). Each slice in the pie should denote the percentage value of the company's total sales.

```python
amount_sold = [6.1,0.9,1.0,0.8,1.2] # creates list of the amount sold of each candy
sizes = [(i/sum(amount_sold)*100) for i in amount_sold] # converts each value of amount_sold to a percentage
sizes # displays list of percentages
```

```python
labels = 'Wonka Bar', 'Everlasting Gobstoppers', 'Fizzy Lifting Drinks', 'Hair Toffee', 'Wonka Gum' # creates pie chart labels
explode = (0, 0, 0, 0.1, 0) # explodes the smallest element of the pie chart out to emphasize it

fig1, ax1 = plt.subplots() # creates a plot that can be made from multiple elements
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90) # uses denoted parameters to create a pie chart
ax1.axis('equal') # ensures pie chart is circular

plt.show() # displays pie chart
```

<b>Question 2:</b> [<span style="color: red;">10 points</span>]<br>
Management wants to see the sales numbers for the years of 1991 to 2021 (included) for (i) Wonka Gum (ii) Fizzy Lifting Drinks and (iii) Hair Toffee. In the file <code>Candy_sales.npz</code> there are 3 different ndarrays (that you can check with <code>.files</code> after using <code>np.load</code>  q2_wg (Wonka Gum), q2_fld (Fizzy Lifting Drinks) and q2_ht (Hair Toffee) for the years in question.<br>

Generate a line plot with a legend and appropriate x and y axis labels. Each line should have a different marker for datapoints.

```python
#Your code goes below
x = np.load('/content/Candy_sales.npz') # loads the data from the Candy_sales.npz file
x.files # displays the data source names within x
```

<b>Question 3:</b> [<span style="color: red;">10 points</span>]<br>
The Willy Wonka Candy company operates in North America, South America and Europe. Management wants to see the breakdown of sales (number of units sold) by region, for each quarter of 2021.<br>

Plot a bar graph for each quarter for each group (NA, SA and EU) with the appropriate plot legends.

```python
year_2021_sales = [10,15,15,21]
#Q1 to Q4 labels
quarters_labels = ['Quarter 1','Quarter 2','Quarter 3','Quarter 4']
#north america sales (in millions)
na_2021_sales = [0.4,0.8,0.8,0.9]
#south america sales
sa_2021_sales = [0.2,0.5,0.3,0.5]
#europe sales
eu_2021_sales = [0.4,0.2,0.4,0.7]
```

```python
#Your code goes below
fig, ax = plt.subplots() # creates a plot that can be made from multiple elements

x = np.arange(len(quarters_labels)) # creates evenly spaced values with the interval given by the amount of elements in quarters_labels
width = 0.25 # value used for bar graph spacing

ax.bar(x - width, na_2021_sales, width, color='red',label='North American Sales') # creates leftmost bar from na_2021_sales with the given parameters
ax.bar(x, sa_2021_sales, width, color='blue', label='South American Sales') # creates middle bar from sa_2021_sales with the given parameters
ax.bar(x + width, eu_2021_sales, width, color='green', label='European Sales') # creates rightmost bar from eu_2021_sales with the given parameters

ax.set_ylabel('Amount of Sales (in millions)') # labels y axis with 'Amount of Sales (in millions)'
ax.set_title('Sales by Quarter on Different Continents') # titles the graph 'Sales by Quarter on Different Continents'
plt.ylim(0,1.2) # sets the range of y values between 0 and 1.2

ax.set_xticks(x) # creates ticks from x interval
ax.set_xticklabels(quarters_labels) # labels x axis with quarters_labels
ax.legend(loc='upper left') # creates a legend in the upper left corner
plt.show() # displays bar graph
```

[BONUS QUESTION] <b>Question 3(a):</b> [<span style="color: red;">10 points</span>]<br>
Before expanding into Africa and Asia, the company wants to see the age demographics of some of their products (i)Wonka Bars (wonka_bars) (ii) Fizzy Lifting Drinks (fld) and (iii) Hair Toffee (hair_toffee). Based on the survey of 10000 users of each product, we have the data on their ages. This is saved in <code>Age_data.npz</code> (See the <code>.files</code> attribute for the various arrays).

Your task is to create histograms based on their ages to see the spread of the distribution. Set number of bins between 25-50 based on your judgement. Each plot must have a legend. You can choose to display it as separate figures or multiple subplots in one figure.



```python
#Your code goes below
x = np.load('/content/Age_data.npz') # loads the data from Age_data.npz file
x.files # displays the data source names within x
```

```python
wonka_bars = x['wonka_bars'] # creates variable containing wonka_bars data
fld = x['fld'] # creates variable containing fld data
hair_toffee = x['hair_toffee'] # creates variable containing hair_toffee data
bins = np.linspace(15, 100, 25) # creates range of data values from near the minimum value to near the maximum value with 25 bina

plt.hist([wonka_bars, fld, hair_toffee], bins, label=['Wonka Bars', 'Fizzy Lifting Drinks', 'Hair Toffee']) # creates histogram with 3 datasets side by side, and given bins and labels
plt.legend(loc='upper right') # creates a legend in the upper right hand corner
plt.show() # displays histogram
```
