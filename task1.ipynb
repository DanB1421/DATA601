The first step to accessing the data within arxiv and gathering metadata is by generating lists of all arxiv papers that fit within each main category of interest, those being Artificial Intelligence (cs.AI), Machine Learning (cs.LG), Computational Complexity (cs.CC), and Hardware Architecture (cs.AR). The generators for the first and second categories produce errors at a certain point, but the lists are still populated with data.

```python
results_generator1 = arxiv.Client(
    page_size=2000, delay_seconds=3, num_retries=3).results(
        arxiv.Search(
            query='cs.AI',
            sort_order=arxiv.SortOrder.Descending,
        )) # generates information on all arxiv papers containing the category cs.AI

cat1 = [] # produces a list
for paper in results_generator1:
    cat1.append(paper) # appends all data for cs.AI papers to the list cat1
```

```python
results_generator2 = arxiv.Client(
    page_size=2000, delay_seconds=3, num_retries=3).results(
        arxiv.Search(
            query='cs.LG',
            sort_order=arxiv.SortOrder.Descending,
        )) # generates information on all arxiv papers containing the category cs.LG

cat2 = [] # produces a list
for paper in results_generator2:
    cat2.append(paper) # appends all data for cs.LG papers to the list cat2
```

```python
results_generator3 = arxiv.Client(
    page_size=2000, delay_seconds=3, num_retries=3).results(
        arxiv.Search(
            query='cs.CC',
            sort_order=arxiv.SortOrder.Descending,
        )) # generates information on all arxiv papers containing the category cs.CC

cat3 = []
for paper in results_generator3:
    cat3.append(paper) # appends all data for cs.CC papers to the list cat3
```

```python
results_generator4 = arxiv.Client(
    page_size=2000, delay_seconds=3, num_retries=3).results(
        arxiv.Search(
            query='cs.AR',
            sort_order=arxiv.SortOrder.Descending,
        )) # generates information on all arxiv papers containing the category cs.AR

cat4 = []
for paper in results_generator4:
    cat4.append(paper) # appends all data for cs.AR papers to the list cat3
```

The next step is to retrieve the metadata from each set and save it into JSON files.

```python
paper_info1_list_unfiltered = [] # generates a list
for paper in cat1: # creates a for loop to generate metadata for each cs.AI paper
    paper_info1 = {} # generates an empty dictionary where metadata will be stored
    paper_info1['title'] = paper.title # adds the title to the dictionary
    paper_info1['authors'] = [] # creates list to be populated with author names in the dictionary
    for author in paper.authors:
        paper_info1['authors'].append(str(author)) # adds author names to the empty list in the dictionary as strings
    paper_info1['summary'] = paper.summary # adds summary to the dictionary
    paper_info1['comment'] = paper.comment # adds comment to the dictionary
    paper_info1['journal_ref'] = paper.journal_ref # adds journal_ref to the dictionary
    paper_info1['doi'] = paper.doi # adds doi to the dictionary
    paper_info1['entry_id'] = paper.entry_id # adds entry_id to the dictionary
    paper_info1['updated'] = str(paper.updated) # adds the most recent updated date to the dictionary as a string
    paper_info1['published'] = str(paper.published) # adds the publication date to the dictionary as a string
    paper_info1['primary_category'] = paper.primary_category # adds the primary category to the dictionary
    paper_info1['categories'] = paper.categories # adds all listed categories to the dictionary
    paper_info1['links'] =  str(paper.links) # adds links to the dictionary as strings
    paper_info1['pdf_url'] = paper.pdf_url # adds pdf_url to the dictionary
    paper_info1_list_unfiltered.append(paper_info1) # adds each paper's dictionary to the list paper_info_1_list_unfiltered

with open('paper_info1_list_unfiltered.json', 'w') as fp:
    json.dump(paper_info1_list_unfiltered, fp) # saves paper_info_1_list_unfiltered as a JSON file
```

The following step is to filter each dataset to ensure that they are all within the correct primary category and a year range between 2017 and 2021. The previously created JSON files are converted into pandas dataframes and filtered by year values in the published column.

```python
df1 = pd.read_json('/content/paper_info1_list_unfiltered.json') # reads the JSON file for the cs.AI papers into a pandas dataframe
df1_filtered = df1[df1['primary_category'] == 'cs.AI'] # creates a filtered dataframe including only papers with cs.AI as the primary category
mask = (df1_filtered['published'] >= '2017-01-01') & (df1_filtered['published'] < '2022-01-01') # creates a filter that only includes papers published from 2017 to 2021
df1_final = df1_filtered.loc[mask] # creates a filtered dataframe using the above filter
df1_final.reset_index(drop=True, inplace=True) # resets the index of the dataframe
df1_final.head() # displays the dataframe's first 5 rows
```
