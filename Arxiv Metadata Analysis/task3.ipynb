```python
import pandas as pd
import matplotlib.pyplot as plt
from ast import literal_eval
```

This task follows a multistep process for each primary category of interest to obtain pie charts. For each dataset, a data frame is created by reading from the previously obtained CSV files. Then, the authors columns are converted from string entries to list entries using the abstract syntax library's literal_eval function. The authors columns are subsequently extracted as lists of author lists. Once extracted, each entry's author list length is determined and placed into a new list to determine how many authors each paper has. The lengths are then categorized into four groups (one author, two authors, three or four authors, and more than four authors) and placed into another new list. Each category is then counted and converted into percentage values by dividing from the length of the category list. These percentages are used to create and display the pie charts for each primary category of interest.

```python
df1_final = pd.read_csv('/content/df1_final.csv', converters={'authors': literal_eval}, index_col=0) # reads the cs.AI CSV from Task 1, treats the first column as an index, and converts the authors column entries into lists
df1_authors_list = df1_final['authors'].to_list() # pulls the author column as a list

author_num_list1 = [] # creates a list
for entry in df1_authors_list:
    num_of_authors = len(entry) # finds the number of authors for each paper
    author_num_list1.append(num_of_authors) # appends the number of authors to author_num_list1

author_num_list1_cat = [] # creates a list
for entry in author_num_list1:
    if entry == 1:
        author_num_list1_cat.append('one author') # appends "one author" to author_num_list1_cat if the author_num_list1 entry is equal to 1
    if entry == 2:
        author_num_list1_cat.append('two authors') # appends "two authors" to author_num_list1_cat if the author_num_list1 entry is equal to 2
    if entry == 3 or entry == 4:
        author_num_list1_cat.append('three or four authors') # appends "three or four authors" to author_num_list1_cat if the author_num_list1 entry is equal to 3 or 4
    if entry > 4:
        author_num_list1_cat.append('more than four authors') # appends "more than four authors" to author_num_list1_cat if the author_num_list1 entry is greater than 4

one_author_count1 = author_num_list1_cat.count('one author') # counts the number of "one author" entries in author_num_list1_cat
two_authors_count1 = author_num_list1_cat.count('two authors') # counts the number of "two authors" entries in author_num_list1_cat
three_or_four_authors_count1 = author_num_list1_cat.count('three or four authors') # counts the number of "three or four authors" entries in author_num_list1_cat
more_than_four_authors_count1 = author_num_list1_cat.count('more than four authors') # counts the number of "more than four authors" entries in author_num_list1_cat
author_num_list1_cat_len = len(author_num_list1_cat) # finds the length of author_num_list1_cat

one_author_size1 = (one_author_count1/author_num_list1_cat_len) * 100 # calculates the percentage of 'one author' entries in the dataset as one_author_size1
two_authors_size1 = (two_authors_count1/author_num_list1_cat_len) * 100 # calculates the percentage of 'two authors' entries in the dataset as two_authors_size1
three_or_four_authors_size1 = (three_or_four_authors_count1/author_num_list1_cat_len) * 100 # calculates the percentage of 'three or four authors' entries in the dataset as three_or_four_authors_size1
more_than_four_authors_size1 = (more_than_four_authors_count1/author_num_list1_cat_len) * 100 # calculates the percentage of 'more than four authors' entries in the dataset as more_than_four_authors_size1

labels = 'Papers with One Author', 'Papers with Two Authors', 'Papers with Three or Four Authors', 'Papers with More than Four Authors' # creates a list of labels for each data category
sizes = [one_author_size1, two_authors_size1, three_or_four_authors_size1, more_than_four_authors_size1] # creates a list of percentages

fig, ax = plt.subplots() # creates a plot that can be made from multiple elements
ax.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90) # uses denoted parameters to create a pie chart
plt.title('Number of Authors for Artificial Intelligence Papers') # titles the chart

plt.show() # displays the pie chart
```
